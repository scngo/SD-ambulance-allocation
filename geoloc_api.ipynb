```python
import numpy as np
import scipy as sp
import pandas as pd
import datetime
import matplotlib.pyplot as plt
```

```python
import random
from geopy.geocoders import Nominatim
import time

geoloc_dict_csv = "geoloc_coord.csv"
dtypes = {
    'Latitude_Decimal': 'float',
    'Longitude_Decimal': 'float',
    'Address': 'str'
}
geoloc_set_read = pd.read_csv(geoloc_dict_csv, dtype=dtypes)
```

```python
geolocator = Nominatim(user_agent="specify_your_app_name_here")
for index, row in geoloc_set_read.iterrows():
    if row['Address'] == ' ':
        try:
            geoloc = str(row['Latitude_Decimal']) + ', ' + str(row['Longitude_Decimal'])
            location = geolocator.reverse(geoloc)
            geoloc_set_read.at[index,'Address'] = location.address
            #print(row['Latitude_Decimal'], row['Longitude_Decimal'], row['Address'])
            print(index, end="\r")
            time.sleep(0.5)
        except:
            time.sleep(1)
```

```python
geoloc_set_read.to_csv("geoloc_coord.csv", index=False, float_format='%.6f')
```
